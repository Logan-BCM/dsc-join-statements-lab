## Select the names of all employees in Boston 

Hint: join the employees and offices tables. Select the first and last name.

```python
# Your code here
pd.read_sql("""SELECT firstName, lastName
            FROM employees
            JOIN offices
            USING (officeCode)
            WHERE city = "Boston"
            """,conn)
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by. Select the office code, city, and number of employees.

```python
# Your code here
pd.read_sql("""SELECT offices.officeCode, offices.city,
            COUNT(employees.employeeNumber) AS num
            FROM offices
            JOIN employees
            USING(officeCode)
            GROUP BY offices.officeCode
            HAVING num = 0;
            """, conn)
```

## Level Up 1: Display the names of every individual product that each employee has sold

Hint: You will need to use multiple `JOIN` clauses to connect all the way from employee names to product names.

```python
# Your code here
pd.read_sql("""SELECT e.lastName,
            e.firstName,
            p.productName
            FROM employees AS e
            JOIN customers AS c
                ON e.employeeNumber = c.salesRepEmployeeNumber
            JOIN orders
                USING(customerNumber)
            JOIN orderdetails
                USING(orderNumber)
            JOIN products AS p
                USING(productCode);
            """,conn)
```

## Level Up 2: Display the number of products each employee has sold

Alphabetize the results by employee last name.

Hint: Use the `quantityOrdered` column from `orderDetails`. Also, think about how to group the data when some employees might have the same first or last name.

```python
# Your code here
pd.read_sql("""SELECT e.lastName,
            e.firstName,
            SUM(od.quantityOrdered) AS num_of_products_sold
            FROM employees AS e
            JOIN customers AS c
                ON e.employeeNumber = c.salesRepEmployeeNumber
            JOIN orders as o
                USING(customerNumber)
            JOIN orderdetails AS od
                USING(orderNumber)
            GROUP BY employeeNumber
            ORDER BY num_of_products_sold DESC;
            
            """,conn)
```

## Level Up 3: Display the names employees who have sold more than 200 different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
# Your code here
pd.read_sql("""SELECT e.lastName, e.firstName,
            COUNT(p.productCode) AS differentProducts
            FROM employees AS e
            JOIN customers
                ON e.employeeNumber = salesRepEmployeeNumber
            JOIN orders
                USING(customerNumber)
            JOIN orderdetails 
                USING(orderNumber)
            JOIN products AS p
                USING(productCode)
            GROUP BY employeeNumber
            HAVING differentProducts > 200
            ORDER BY differentProducts DESC;
            """,conn)
```
